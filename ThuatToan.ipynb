```python
# ------------------ BÀI 3 ---------------------

import math

# ------------------ Read Matrix ---------------------
def read_cost_matrix(filename):
    G = []
    with open(filename, "r") as f:
        for line in f:
            row = list(map(int, line.split()))
            G.append(row)
    return G

# ------------------ TSP Branch and Bound ---------------------
best_cost = math.inf
best_path = []

def tsp_branch_bound(G, visited, current_city, count, cost, path):
    global best_cost, best_path
    n = len(G)

    # Nếu đã đi hết n thành phố
    if count == n and G[current_city][0] > 0:
        total_cost = cost + G[current_city][0]
        if total_cost < best_cost:
            best_cost = total_cost
            best_path = path + [0]
        return

    # Nhánh cận
    for next_city in range(n):
        if not visited[next_city] and G[current_city][next_city] > 0:
            new_cost = cost + G[current_city][next_city]

            # Cắt nhánh nếu chi phí đã vượt
            if new_cost >= best_cost:
                continue

            visited[next_city] = True
            tsp_branch_bound(G, visited, next_city, count + 1, new_cost, path + [next_city])
            visited[next_city] = False

# ------------------ Main Function ---------------------
def solve_tsp(filename):
    global best_cost, best_path
    G = read_cost_matrix(filename)
    n = len(G)

    visited = [False] * n
    visited[0] = True  # bắt đầu từ thành phố 0

    tsp_branch_bound(G, visited, 0, 1, 0, [0])

    return best_path, best_cost


# ------------------ RUN ---------------------
path, cost = solve_tsp("tsp.txt")

print("Chu trình tối ưu:")
print(" → ".join("City"+str(c) for c in path))

print("\nChi phí nhỏ nhất:", cost)
```
