```python
# ------------------ BÀI 2 ---------------------

def read_adj_matrix(filename):
    G = []
    with open(filename, 'r') as f:
        for line in f:
            row = list(map(int, line.split()))
            G.append(row)
    return G

def calculate_degrees(G):
    return [sum(row) for row in G]

def sort_nodes_by_degree(node, degree):
    sorted_nodes = []
    used = []
    n = len(node)

    for _ in range(n):
        max_deg = -1
        idx = -1
        for i in range(n):
            if i not in used and degree[i] > max_deg:
                max_deg = degree[i]
                idx = i
        used.append(idx)
        sorted_nodes.append(node[idx])

    return sorted_nodes

def color_graph(G):
    n = len(G)
    node = [chr(ord('A') + i) for i in range(n)]
    degree = calculate_degrees(G)
    sorted_nodes = sort_nodes_by_degree(node, degree)
    index_map = {node[i]: i for i in range(n)}

    colorDict = {node[i]: ["Blue", "Red", "Yellow", "Green"] for i in range(n)}
    solution = {}

    for v in sorted_nodes:
        chosen = colorDict[v][0]
        solution[v] = chosen

        # Loại màu khỏi đỉnh kề
        for j in range(n):
            if G[index_map[v]][j] == 1:
                if chosen in colorDict[node[j]]:
                    colorDict[node[j]].remove(chosen)

    return solution


def print_coloring(solution):
    print("\n--- KẾT QUẢ TÔ MÀU ---")
    for v, c in sorted(solution.items()):
        print(f"{v} → {c}")


# ------------------ MAIN ---------------------

file = "graph.txt"
G = read_adj_matrix(file)
solution = color_graph(G)
print_coloring(solution)
```
